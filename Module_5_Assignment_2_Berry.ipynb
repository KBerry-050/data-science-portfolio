# MODELING

```python
# modeling
X = df_new.drop('Bankrupt',axis=1)
y = df_new['Bankrupt']
```

```python
# define train and test sets
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20)
```

```python
# create the model
CLF1 = RandomForestClassifier(n_estimators=100)
```

```python
# train the model and make predictions
CLF1.fit(X_train, y_train)

y_pred = CLF1.predict(X_test)
```

```python
# check accuracy
print("The accuracy score for this Random Forest Classifier is: ", accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
# check feature importance
features = pd.Series(CLF1.feature_importances_, index=X.columns).sort_values(ascending=False)
features
```

```python
# visualize importance for top 10 features
TOP10 = features.head(10)
sns.barplot(x=TOP10, y=TOP10.index)
plt.xlabel("Feature Importance Score")
plt.ylabel("Features")
plt.title("Random Forest Classifier - Important Features")
plt.show()
```

```python
# create function to visualize feature importance
def feat_import(x,y):
    sns.barplot(x=x, y=y.index)
    plt.xlabel("Feature Importance Score")
    plt.ylabel("Features")
    plt.title("This Model's Important Features")
    plt.show()
```

```python
feat_import(TOP10, TOP10)
```

```python
# hyperparameter tuning - create a parameter grid

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]

# Number of features to consider at every split
max_features = ['auto', 'sqrt']

# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)

# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]

# Method of selecting samples for training each tree
bootstrap = [True, False]
```

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}

print(random_grid)
```

```python
# use random grid to search for the best parameters
CLF2 = RandomForestClassifier()

CLF2_random = RandomizedSearchCV(estimator=CLF2, param_distributions=random_grid, n_iter=100, cv=3, verbose=2, random_state=42, n_jobs=-1)
```

```python
# fit the model and find the best parameters
CLF2_random.fit(X_train, y_train)
```

```python
# view the best parameters
CLF2_random.best_params_
```

```python
# check
best_random = CLF2_random.best_estimator_
best_random
```

```python
# fit model again with the best hyperparamters
CLF2_random.best_estimator_.fit(X_train,y_train)

# make predictions and get accuracy
y_pred_random = CLF2_random.best_estimator_.predict(X_test)

accuracy_score(y_test,y_pred_random)
```

```python
# create function to compare accuracy
def evaluate(model, test_features, test_labels):
    predictions = model.predict(test_features).astype('int')
    errors = abs(predictions - test_labels)
    mape = 100 * np.mean(errors / test_labels)
    accuracy = 100 - mape
    print('Model Performance')
    print('Average Error: {:0.4f} degrees.'.format(np.mean(errors)))
    print('Accuracy = {:0.2f}%.'.format(accuracy))
    
    return accuracy
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score, RepeatedStratifiedKFold
from numpy import mean,std
```

```python
# define the model
GB_model = GradientBoostingClassifier()

# define the evaluation method
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
```

```python
# evaluate the model on the dataset
GB_scores = cross_val_score(GB_model, X_train, y_train, scoring='accuracy', cv=cv, n_jobs=-1)

# check performance across the 10 scores
print('Mean Accuracy: %.3f (%.3f)' % (mean(GB_scores), std(GB_scores)))
```

```python
# check model performance for against the test set
GB_model.fit(X_train,y_train)

y_pred_GB = GB_model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_GB)
```

```python
# perform hyperparameter tuning for the Gradient Boosted model

# define grid of values
grid = dict()
grid['n_estimators'] = [10, 50, 100, 500]
grid['learning_rate'] = [0.0001, 0.001, 0.01, 0.1, 1.0]
grid['subsample'] = [0.5, 0.7, 1.0]
grid['max_depth'] = [3, 7, 9]

# define evaluation procedure
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)

# define grid search procedure
grid_search = GridSearchCV(estimator=GB_model, param_grid=grid, n_jobs=-1, cv=cv, scoring='accuracy')
```

```python
# execute the grid search
grid_result = grid_search.fit(X_train, y_train)
# summarize the best score and configuration
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
# summarize all scores that were evaluated
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```
